```python
import matplotlib.pyplot as plt
from math import *
```

```python
def Func1(x):
    f=1./(1+x**2)
    return f

def Func2(x):
    f=x**(1/3)*exp(sin(x))
    return f
```

```python
prec = 10e-8
```

```python
l=[]
k = 1
print("Intervals, trapeze 1")
while True:
    if len(l)>1:
      if (l[-1]-l[-2])/3 < prec:
        break
    s=0
    k=2*k
    dx=1/k
    for i in range(k):
        s=s+(Func1(i*dx)+Func1((i+1)*dx))*dx/2
    l.append(2*s)
    print(k,":", 2*s)
d = [abs(item - l[-1]) for item in l]
plt.yscale('log')
plt.plot(d)
plt.show()
diff = log((d[0]/d[-2]), 2)/log(k/4,2) #pre-last diff - first by degrees difference
print(diff)
l1t = l
i1t = log(k,2)
```

```python
l=[]
k = 1
print("Intervals, simpson 1")
while True:
    if len(l)>1:
      if (l[-1]-l[-2])/15 < prec:
        break
    s=0
    k=2*k
    dx=1/k
    for i in range(k):
        s=s+(Func1(i*dx)+Func1((i+1)*dx)+4*Func1((i+0.5)*dx))*dx/6
    l.append(s)
    print(k,":", 2*s)
d = [abs(item - l[-1]) for item in l]
plt.yscale('log')
plt.plot(d)
plt.show()
diff = log((d[0]/d[-2]), 2)/log(k/4,2) #pre-last diff - first by degrees difference
print(diff)
l1s = l
i1s = log(k,2)
```

```python
l=[]
k = 1
print("Intervals, simpson 1")
while True:
    if len(l)>1:
      if (l[-1]-l[-2])/15 < prec:
        break
    s=0
    k=2*k
    dx=1/k
    for i in range(k):
        s=s+(Func2(i*dx)+Func2((i+1)*dx))*dx/2
    l.append(s)
    print(k,":", 2*s)
d = [abs(item - l[-1]) for item in l]
plt.yscale('log')
plt.plot(d)
plt.show()
diff = log((d[0]/d[-2]), 2)/log(k/4,2) #pre-last diff - first by degrees difference
print(diff)
l2t = l
i2t = log(k,2)
```
